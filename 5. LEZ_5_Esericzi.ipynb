# 4. Rispondere alle domande
Trasforma le seguenti richiesti in formule per estrarre infromazioni dal dataframe
1. Mostrami le vendite maggiori di 10

2. Mostrami i dati del 2018

3. Mostrami le vendite maggiori di 13 e l'anno è il 2018

4. Mostrami tutto TRANNE i casi in cui le vendite sono maggiori di 13 e l'anno è il 2018

5. Mostrami i dati dove le vendite divise per 3 sono maggiori di 3

6. Mostrami i dipendenti i cui nomi sono alfabeticamente dopo la J



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales = pd.DataFrame(
    data={
        "employee": [
            "Katrina",
            "Guanyu",
            "Jan",
            "Roman",
            "Jacqueline",
            "Paola",
            "Esperanza",
            "Alaina",
            "Egweyn",
        ],
        "sales": [14, 17, 6, 12, 8, 3, 7, 15, 5],
        "year": [2018, 2019, 2020, 2018, 2020, 2019, 2019, 2020, 2020],
    }
)
sales
```

***Punto 1: Mostrami le vendite maggiori di 10***

```python
# OBIETTIVO: Filtrare il DF per mostrare solo le righe dove la colonna sales è MAGGIORE di 10.

print("Le righe con vendite maggiori di 10 sono:")
sales[sales["sales"] > 10]
```

***Punto 2: Mostrami i dati del 2018***

```python
# OBIETTIVO: Selezionare solo le righe dove la colonna years è UGUALE a 2018.
 
print("I dati del 2018 sono:")
sales[sales["year"] == 2018]
```

***Punto 3:  Mostrami le vendite maggiori di 13 e l'anno è il 2018***

```python
# OBIETTIVO: Filtrare le righe dove ENTRAMBE le condizioni SONO VERE.

print("Vendite maggiori di 13 nel 2018:")
print(sales[(sales["sales"] > 13) & (sales["year"] == 2018)])

## INFO: In Pandas, per combinare le condizioni bisogna usare:
#        - & per "E LOGICO"
#        - | per "OPPURE"
#        - Le condizioni vanno sempre tra le PARENTESI TONDE ()
```

***Punto 4: Mostrami tutto TRANNE i casi in cui le vendite sono maggiori di 13 e l'anno è il 2018***

```python
# OBIETTIVO: ESCLUDERE le righe che soddisfanbo entrambe le condizioni:     -sales > 13    e   -years == 2018

print(" Tutti i dati tranne quelli con vendite > 13 e anno = 2018:")
print(sales[~((sales["sales"] > 13) & (sales["year"] == 2018))])

## INFO:   - & è una condizione COMBINATA
#          - ~(...): NEGA la condizione. Tutto cio che non soddisfa sales > 13 e years == 2018. 
```

***Punto 5: Mostrami i dati dove le vendite divise per 3 sono maggiori di 3***

```python
# OBIETTIVO: Filtrare le righe dove:    (sales / 3) > 3

print("Dati dove le vendite divise per 3 sono maggiori di 3:")
print(sales[(sales["sales"] / 3) > 3])

## INFO: Questo include tutte le righe dove la colonna sales 
#        è MAGGIORE di 9, dato che: (sales / 3) > 3  ⟺  sales > 9
```

***Punto 6: Mostrami i dipendenti i cui nomi sono alfabeticamente dopo la J***

```python
# OBIETTIVO: Filtrare i dipendenti il cui nome inizia con una lettera DOPO la J (cioè K,L,M ... Z).

print("Dipendenti con nome dopo la lettera 'J':")
print(sales[sales["employee"] > "J"])

## INFO: In python, le STRINGHE possono essere confrontate alfabeticamente.
#       "Katrina" > "J" → ✅ True
#       "Esperanza" > "J" → ❌ False (perché "E" viene prima)
```

# 5. Analisi di un dataset di video giochi
1. Quanti videogiochi sono stati pubblicati
2. Quali sono i generi più popolari? Mostralo con un bar plot
3. Plotta l'evoluzione del numero di giochi pubblicati nel tempo
4. Prova a riprodurre questo plot 



```python
# Import dei pacchetti
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caricamento del dataset
url = 'https://zenodo.org/record/5898311/files/vgsales.csv'
df = pd.read_csv(url)
df.head()
```

***Punto 1: Quanti videogiochi sono stati pubblicati?***

```python
## OBIETTIVO: Contare il numero totale di righe del DS (ogni riga rappresenta un videogioco)

print(" Numero totale di videogiochi pubblicati:")
print(len(df))
```

***Punto 2: Quali sono i generi più popolari? Mostralo con un bar plot***

```python
## OBIETTIVO: Contare quanti giochi appartengono a ciascun genere  e   visualizzarli con un grafico a barre.

# Conteggio dei giochi per genere
generi = df['Genre'].value_counts()

# Dimensioni del grafico
plt.figure(figsize=(10, 5))

# Creo il grafico
sns.barplot(
    x=generi.index, 
    y=generi.values, 
    hue=generi.index, 
    palette='Set2', 
    legend=False)

# Aggiungo il titolo e le etichette agli assi 
plt.title(" Numero di videogiochi per genere")
plt.xlabel("Genere")
plt.ylabel("Numero di giochi")

# Ruoto le etichette sull'asse x per renderle leggibili
plt.xticks(rotation=45)

# Miglioro il layout per evitare sovrapposizioni
plt.tight_layout()

# Mostro il grafico
plt.show()
```

***Punto 3: Plotta l’evoluzione del numero di giochi pubblicati nel tempo***

```python
## OBIETTIVO: Vedere quanti giochi sono stati pubblicati per anno (Year) e mostrarlo con una line plot

# Calcolo il numero di giochi pubblicati per ogni anno
giochi_per_anno = df['Year'].value_counts().sort_index()  # Ordino gli anni in ordine crescente

# Dimensioni del grafico
plt.figure(figsize=(10, 5))  

# Creo il grafico
sns.lineplot(
    x=giochi_per_anno.index, 
    y=giochi_per_anno.values, 
    marker='o')

#  Aggiungo il titolo e le etichette agli assi
plt.title("Giochi pubblicati per anno")
plt.xlabel("Anno")
plt.ylabel("Numero di giochi")

# Miglioro il layout
plt.tight_layout()

# Mostro il grafico
plt.show()
```

***Punto 4: Riproduci il plot***

```python
## OBIETTIVO: Riprodurre il grafico che è all'inizio dell'esercizio.

# Seleziono solo i 5 generi che compaiono nel grafico originale
generi_scelti = ["Action", "Misc", "Role-Playing", "Shooter", "Sports"]

# Raggruppo e sommo solo per quei generi
vendite_filtrate = df[df["Genre"].isin(generi_scelti)].groupby("Genre")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()

# Riordino i generi come nel grafico originale
ordine_originale = ["Action", "Misc", "Role-Playing", "Shooter", "Sports"]
vendite_filtrate = vendite_filtrate.loc[ordine_originale]

# Creo il grafico a barre impilate
vendite_filtrate.plot(
    kind="bar",
    stacked=True,
    figsize=(10, 6),
    color=["blue", "orange", "green", "red"]  
)

# Aggiungo il titolo e le etichette per agli assi
plt.title("Vendite per regione per genere")
plt.xlabel("Genere")
plt.ylabel("Milioni di unità")

# Ruoto le etichette sull'asse x per renderle leggibili
plt.xticks(rotation=45)

# Legenda in alto centrata
plt.legend(title="Regione", loc="upper center", bbox_to_anchor=(0.5, 0.99))


# Miglioro il layout
plt.tight_layout()

# Mostro il grafico
plt.show()
```
